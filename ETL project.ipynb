```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
ca_videos = ("resources/CAvideos.csv")
ca_videos_df = pd.read_csv(ca_videos)
ca_videos_df.head()
```

```python
gb_videos = ("resources/GBvideos.csv")
gb_videos_df = pd.read_csv(gb_videos)
gb_videos_df.head()
```

```python
us_videos = ("resources/USvideos.csv")
us_videos_df = pd.read_csv(us_videos)
us_videos_df.head()
```

### Transform premise DataFrame

```python
# Create a filtered dataframe from specific columns
ca_videos_cols = ["video_id", "trending_date", "title","category_id","publish_time","views","likes","dislikes","comment_count"]
ca_videos_transformed= ca_videos_df[ca_videos_cols].copy()

# Rename the column headers
ca_videos_transformed = ca_videos_transformed.rename(columns=
                                                             {"video_id": "id",
                                                              "trending_date": "trendingdate",
                                                              "title": "title",
                                                              "category_id": "categoryid",
                                                              "publish_time": "publisheddate",
                                                              "views": "views",
                                                              "likes": "likes",
                                                              "dislikes": "dislikes",
                                                              "comment_count": "commentcount"})

# Clean the data by dropping duplicates and setting the index
ca_videos_transformed.drop_duplicates("id", inplace=True)

ca_videos_transformed.set_index('id', inplace=True)


ca_videos_transformed.head()
# ca_videos_transformed_renamed.columns

```

### Transform county DataFrame

```python
gb_videos_cols = ["video_id", "trending_date", "title","category_id","publish_time","views","likes","dislikes","comment_count"]
gb_videos_transformed = gb_videos_df[gb_videos_cols].copy()

# Rename the column headers
gb_videos_transformed = gb_videos_transformed.rename(columns=
                                                              {"video_id": "id",
                                                              "trending_date": "trendingdate",
                                                              "title": "title",
                                                              "category_id": "categoryid",
                                                              "publish_time": "publisheddate",
                                                              "views": "views",
                                                              "likes": "likes",
                                                              "dislikes": "dislikes",
                                                              "comment_count": "commentcount"})
# Set index
gb_videos_transformed.drop_duplicates("id", inplace=True)

gb_videos_transformed.set_index("id", inplace=True)

gb_videos_transformed.head()
```

```python
us_videos_cols = ["video_id", "trending_date", "title","category_id","publish_time","views","likes","dislikes","comment_count"]
us_videos_transformed = us_videos_df[us_videos_cols].copy()

# Rename the column headers
us_videos_transformed = us_videos_transformed.rename(columns=
                                                              {"video_id": "id",
                                                              "trending_date": "trendingdate",
                                                              "title": "title",
                                                              "category_id": "categoryid",
                                                              "publish_time": "publisheddate",
                                                              "views": "views",
                                                              "likes": "likes",
                                                              "dislikes": "dislikes",
                                                              "comment_count": "commentcount"})
# Set index
us_videos_transformed.drop_duplicates("id", inplace=True)

us_videos_transformed.set_index("id", inplace=True)

us_videos_transformed.head()
# us_videos_transformed.columns
```

### Create database connection

```python
connection_string = "postgres:@naonao@localhost:5432/YouTube_Video_DB"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
ca_videos_transformed.to_sql(name='cavideos', con=engine, if_exists='append', index=True)
```

```python
gb_videos_transformed.to_sql(name='gbvideos', con=engine, if_exists='append', index=True)
```

```python
us_videos_transformed.to_sql(name='usvideos', con=engine, if_exists='append', index=True)
```
